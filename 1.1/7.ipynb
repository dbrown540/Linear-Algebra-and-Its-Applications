# How to plot planes in python

### 1. Import packages numpy, matplotlib.pyplot, and mpl_toolkits.mplot3d.

#### NumPy: numerical operations
#### Matplotlib: For plotting
#### Axes3D: Module for creating 3D plots

```python
import numpy as np
import matplotlib.pyplot as plt
```

### 2. Define constants for the plane equation

```python
a1, b1, c1, d1  = 1, 7, 3, -4
a2, b2, c2, d2 = 0, 1, -1, 3
a3, b3, c3, d3 = 0, 0, 1, -2
a4, b4, c4, d4 = 0, 0, 0, 1
```

### 3. Create a x-y space and a meshgrid to generate all possible (x, y) values in a 3D plot

```python
x1 = np.linspace(-10, 10, 100)
y1 = np.linspace(-10, 10, 100)
xx1, yy1 = np.meshgrid(x1, y1)

x2 = np.linspace(-10, 10, 100)
y2 = np.linspace(-10, 10, 100)
xx2, yy2 = np.meshgrid(x2, y2)

x3 = np.linspace(-10, 10, 100)
y3 = np.linspace(-10, 10, 100)
xx3, yy3 = np.meshgrid(x3, y3)

x4 = np.linspace(-10, 10, 100)
y4 = np.linspace(-10, 10, 100)
xx4, yy4 = np.meshgrid(x4, y4)
```

### 4. Calculate the z values

```python
# Given the equation for a plane: ax + by + cz = d, solve for z

z1 = (d1 - a1 * xx1 - b1 * yy1) / c1
# print(xx.shape, yy.shape, z.shape) -> Notice how these are all the same shape.. this is because of meshgrid

z2 = (d2 - a2 * xx2 - b2 * yy2) / c2
z3 = (d3 - a3 * xx3 - b3 * yy3) / c3
z4 = (d4 - a4 * xx4 - b4 * yy4) / c4
```

### 5. Create a 3D plot

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
# Set the viewing angle
ax.view_init(elev=20, azim=30)  # Adjust elev and azim as needed
ax.plot_surface(xx1, yy1, z1, alpha=0.5, color='b', label = 'Plane 1')
ax.plot_surface(xx2, yy2, z2, alpha=0.5, color='r', label = 'Plane 2')
ax.plot_surface(xx3, yy3, z3, alpha=0.5, color='g', label = 'Plane 3')
ax.plot_surface(xx4, yy4, z4, alpha=0.5, color='y', label = 'Plane 4')

# Create a legend using Line3D objects
legend1_symbol = plt.Line2D([0], [0], linestyle="none", c='b', marker='o')
legend2_symbol = plt.Line2D([0], [0], linestyle="none", c='r', marker='o')
legend3_symbol = plt.Line2D([0], [0], linestyle="none", c='g', marker='o')
legend4_symbol = plt.Line2D([0], [0], linestyle="none", c='y', marker='o')

# Add legend to the figure
ax.legend([legend1_symbol, legend2_symbol, legend3_symbol, legend4_symbol],['Plane 1', 'Plane 2', 'Plane 3', 'Plane 4'])

# Set the labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Set the title
ax.set_title('Problem 7')

# Show the plot
plt.show()
```
